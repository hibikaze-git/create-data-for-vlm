```python
def calculate_image_hash(image):
    # 
    image_bytes = image.tobytes()
    
    # SHA-256
    hash_md5 = hashlib.sha256(image_bytes).hexdigest()
    return hash_md5
```

```python
image_hash = calculate_image_hash(dataset["train"][0]["images"][0])
```
